```python
columnas_eliminar=['sex_','unnamed:_0_x','unnamed:_0_y','href']
```

```python
#eliminamos las columnas que habíamos arreglado en pandas

df.drop(columnas_eliminar, axis=1, inplace=True)
df.head(2)
```

```python
# ¿Hay valores duplicados en nuestro dataframe? En caso de que los haya, eliminándolos.

df.duplicated().sum()
```

```python
# Como hemos visto, algunas columnas no tienen el tipo de datos que deberían. 
# Cambiad el tipo de dato para la columna de year.

df.info() #vemos el tipo de dato de la columna year que es un float
```

```python
df['year']=df['year'].astype('int64', errors='ignore')
```

```python
df.info() #verificamos que el cambio se ha realizado correctamente
```

```python
# age
# ❌ esta en formato string cuando debería ser integer y en algunos casos tenemos rangos de edad

df['age']=df['age'].astype('int64', errors='ignore')
df.info()
```

```python
# En la columna de country poner todos los valores en minúsculas.

def minusculas(x):
    try:
        if isinstance(x,str): #buscado en pep8
                return x.lower()
        else:
            return x
    except:
        return x
```

```python
df['country']= df['country'].apply(minusculas)
df.head()
```
