Hipótesis
Las hipótesis que os planteamos son:



¿Cuál es la relación entre la especie y el tipo de ataque (si es fatal o no)?
¿Cómo han evolucionado los ataques a lo largo del tiempo?


¿Es Australia el sitio más peligroso y letal para estar relajada en la playa?

```python
paises_letal=df.groupby(['fatal_bien','Country']).size().reset_index().sort_values(by=0, ascending=False)
paises_letal_y=paises_letal[paises_letal['fatal_bien']=='Y'].sort_values(by=0, ascending=False)
paises_letal_y

#No es el más peligroso pero tampoco el mas más seguro porque es la posición número 2
```

¿Hay diferencias entre los países en los ataques a hombres y mujeres?

```python
df_generos_pais=df.groupby(['genero_bien','Country']).size().reset_index().sort_values(by=0, ascending=False)
df_generos_pais

#Podemos observar que el género másculino es el que más ataques registra en los 3 países con más ataques.
```

¿Cuáles son las edades que más sufren ataques?

```python
df_edades=df.groupby(['Age']).size().reset_index().sort_values(by=0, ascending=False)
df_edades

#Según se muestra, la edad de 19 años es la que más ataques tiene registrados seguido de 18 y 20 años.
```

Independientemente de la edad, ¿sufren los hombres más ataques que las mujeres?

```python
df_genero=df.groupby(['genero_bien']).size().reset_index().sort_values(by=0, ascending=False)
df_genero

#El género masculino es definitivamente el que más ataques sufre.
```

¿En qué mes ocurren más ataques?

```python
df_mes=df.groupby(['Mes']).size().reset_index().sort_values(by=0, ascending=False)
df_mes

#El mes de Agosto cuenta con más registros.
```

¿Cuál es la relación entre la especie y el tipo de ataque (si es fatal o no)?

```python
df_especie_fatal=df.groupby(['Species','fatal_bien']).size().reset_index().sort_values(by=0, ascending=False)
df_especie_fatal

#Podemos observar que la especie White es la que más ataques registra pero no son fatales, pero, 
# dentro de esos mismos la especie que más ataques fatales registra es la misma, el White.
```

```python
df_generos=df.groupby(['genero_bien','Country','fatal_bien']).size().reset_index()
df_generos

#El género con más ataques es el masculino
```

```python
df_generos=df_generos[df_generos['fatal_bien']=='Y'].sort_values(by=0, ascending=False)
df_generos

#El género con más registros fatales de ataques es también el masculino.
```
